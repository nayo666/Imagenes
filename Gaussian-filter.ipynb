```python
import numpy as np
#se pide el tama√±o de la matriz
n=5
#se define una matriz de zeros de nxn
a=np.zeros(shape=(n,n))
#se define sigma (dev estandar)
sigx=1
sigy=1

for u in range(0,n):
    for v in range(0,n):
        dx=(int(n/2)-u)
        dy=(int(n/2)-v)
        a[u,v]=np.exp( -( ((dx**2)/(2*sigx**2))+((dy**2)/(2*sigy**2)) ) )
total=np.sum(a)
a=a/total
print(np.sum(a))
```

```python
with np.printoptions(threshold=np.inf):
    print(a)
```

```python
import cv2
import skimage as sk
from skimage import data, io, filters
import warnings
from IPython.display import Image
warnings.filterwarnings("ignore")
img = cv2.imread('lena.jpg',0)
# img=sk.io.imread('lena.jpg')
```

```python
sk.io.imshow(img)
```

```python
L=img.shape
x=L[0]
y=L[1]
smoth=np.zeros(shape=(x,y))
for j in range(0,x):
    for i in range(0,y):
        for u in range(0,n):
            for v in range(0,n):
                d1=(int(n/2)-u)
                d2=(int(n/2)-v)
                if (d1+j)>=0 and (d2+i)>=0 and (d1+j)<x and (d2+i)<y:
                    smoth[j,i]=smoth[j,i]+a[d1,d2]*img[d1+j,d2+i]
maxi=np.amax(smoth)
for j in range(0,x):
    for i in range(0,y):
        smoth[j,i]=(smoth[j,i]/maxi)*256
```

```python
sk.io.imshow(smoth, cmap="gray")
```
